## **Getting started**
</div>

The document you are reading is not a static web page, but an interactive environment called a **Colab notebook** that lets you write and execute code.

For example, here is a **code cell** with a short Python script that computes a value, stores it in a variable, and prints the result:

```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

To execute the code in the above cell, select it with a click and then either press the play button to the left of the code, or use the keyboard shortcut "Command/Ctrl+Enter". To edit the code, just click the cell and start editing.

Variables that you define in one cell can later be used in other cells:

```
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

Colab notebooks allow you to combine **executable code** and **rich text** in a single document, along with **images**, **HTML**, **LaTeX** and more. When you create your own Colab notebooks, they are stored in your Google Drive account. You can easily share your Colab notebooks with co-workers or friends, allowing them to comment on your notebooks or even edit them. ## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Load the consumer complaint dataset (replace 'complaints.csv' with your dataset)
df_full = pd.read_csv('/content/drive/My Drive/complaints.csv')

# Select relevant columns
df = df_full[['Consumer complaint narrative', 'Product']].copy()

# Remove rows with missing data
df.dropna(inplace=True)

# Select a smaller portion of the data for demonstration purposes (optional)
df = df.head(1000)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['Consumer complaint narrative'], df['Product'], test_size=0.25, random_state=0)

# Create a label encoder to convert target labels to numerical values
encoder = LabelEncoder()
y_train_num = encoder.fit_transform(y_train)

# Define a function to build and evaluate a model using a pipeline
def build_and_evaluate_pipeline(model, model_name):
    # Create a text classification pipeline
    text_clf = Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english')),
        ('clf', model)
    ])

    # Fit the pipeline on the training data
    text_clf.fit(X_train, y_train)

    # Make predictions on the test data
    preds = text_clf.predict(X_test)

    # Inverse transform predictions to get category names

    # Calculate accuracy
    accuracy = accuracy_score(y_test, preds)
    print(f'{model_name} Accuracy: {accuracy * 100:.2f}%')

    # Generate classification report
    class_report = classification_report(y_test, preds, output_dict=True)
    print(f'{model_name} Classification Report:\n{classification_report(y_test, preds)}')

    # Generate confusion matrix
    conf_matrix = confusion_matrix(y_test, preds)
    print(f'{model_name} Confusion Matrix:\n{conf_matrix}')

    # Visualize the classification report
    plt.figure(figsize=(8, 6))
    sns.heatmap(pd.DataFrame(class_report).iloc[:-1, :].T, annot=True, cmap='Blues')
    plt.title(f'{model_name} Classification Report')
    plt.show()

    # Visualize the confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
    plt.title(f'{model_name} Confusion Matrix')
    plt.show()

    # Create a DataFrame for comparison
    comparison_df = pd.DataFrame({'ID': range(len(X_test)),
                                  'Complaint': X_test.values,
                                  'Actual Product': y_test.values,
                                  'Prediction': preds})

    return accuracy, comparison_df

# Create and evaluate the Random Forest model
rf_accuracy, rf_comparison_df = build_and_evaluate_pipeline(RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1), 'Random Forest')

# Create and evaluate the Support Vector Machines (SVM) model
svm_accuracy, svm_comparison_df = build_and_evaluate_pipeline(SVC(kernel='linear', C=1, probability=True, random_state=0), 'Support Vector Machines (SVM)')

# Create and evaluate the Gradient Boosting model
gb_accuracy, gb_comparison_df = build_and_evaluate_pipeline(GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=0), 'Gradient Boosting')

# Compare accuracy using a bar chart
models = ['Random Forest', 'SVM', 'Gradient Boosting']
accuracies = [rf_accuracy, svm_accuracy, gb_accuracy]

plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color=['blue', 'green', 'orange'])
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Comparison of Model Accuracies')
plt.ylim(0.0, 1.0)
plt.show()

# Display the comparison DataFrame for each model
print('Random Forest Comparison:')
print(rf_comparison_df.head())

print('SVM Comparison:')
print(svm_comparison_df.head())

print('Gradient Boosting Comparison:')
print(gb_comparison_df.head())

# You can further fine-tune the models and explore additional options for optimization.
```
